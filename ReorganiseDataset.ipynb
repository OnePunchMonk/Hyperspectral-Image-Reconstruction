# Approach to organise test , train, val from requisite datasets

## Tomatao,Apple,Kiwi Seletcted for training

## Blueberry,Strawberry selected for testing

```python
import os
import shutil

# Define the base paths for the datasets and the new structure
base_path = r"C:\Users\aggar\Downloads\datasets\datasets"  
output_path = r"C:\Users\aggar\Downloads\datasets_reorganised_new"

# Define datasets for training/validation and testing
train_val_datasets = ['dataset_tomato', 'dataset_apple', 'dataset_kiwi']
test_datasets = ['dataset_strawberries', 'dataset_blueberries']

# Function to create the directory structure
def create_dir_structure(base_path, categories):
    for category in categories:
        category_path = os.path.join(base_path, category)
        if not os.path.exists(category_path):
            os.makedirs(category_path)

# Function to move files to the new structure
def move_files(dataset, category, output_path):
    dataset_path = os.path.join(base_path, dataset)
    reconstruction_path = os.path.join(dataset_path, 'reconstruction')

    # Read split files
    with open(os.path.join(reconstruction_path, 'split_txt', 'train_list.txt'), 'r') as f:
        train_files = f.read().splitlines()
    with open(os.path.join(reconstruction_path, 'split_txt', 'valid_list.txt'), 'r') as f:
        val_files = f.read().splitlines()

    with open(os.path.join(reconstruction_path, 'split_txt', 'test_list.txt'), 'r') as f:
        test_files = f.read().splitlines()
#If dataset belongs to train_val_datasets list add images to train,val folder
    if dataset in train_val_datasets:
        # Move training files directly to category folder
        for file in train_files:
            mat_file = os.path.join(reconstruction_path, 'Train_Spec', f'{file}.mat')
            rgb_file = os.path.join(reconstruction_path, 'Train_RGB', f'{file}_RGB_D_gc.png')
            dest_folder = os.path.join(output_path, 'training', file)  
            os.makedirs(dest_folder, exist_ok=True)
            shutil.copy(mat_file, os.path.join(dest_folder, f'{file}.mat'))
            shutil.copy(rgb_file, os.path.join(dest_folder, f'{file}_RGB_D_gc.png'))

        # Move validation files directly to category folder
        for file in val_files:
            mat_file = os.path.join(reconstruction_path, 'Train_Spec', f'{file}.mat')
            rgb_file = os.path.join(reconstruction_path, 'Train_RGB', f'{file}_RGB_D_gc.png')
            dest_folder = os.path.join(output_path, 'validation', file) 
            os.makedirs(dest_folder, exist_ok=True)
            shutil.copy(mat_file, os.path.join(dest_folder, f'{file}.mat'))
            shutil.copy(rgb_file, os.path.join(dest_folder, f'{file}_RGB_D_gc.png'))
    else:

        for file in test_files:
            mat_file = os.path.join(reconstruction_path, 'Train_Spec', f'{file}.mat')
            rgb_file = os.path.join(reconstruction_path, 'Train_RGB', f'{file}_RGB_D_gc.png')
            dest_folder = os.path.join(output_path, 'testing', file)  # Notice the change here
            os.makedirs(dest_folder, exist_ok=True)
            shutil.copy(mat_file, os.path.join(dest_folder, f'{file}.mat'))
            shutil.copy(rgb_file, os.path.join(dest_folder, f'{file}_RGB_D_gc.png'))

# Create the new directory structure
create_dir_structure(output_path, ['training', 'validation', 'testing'])

# Move files for training/validation datasets
for dataset in train_val_datasets:
    move_files(dataset, dataset.split('_')[-1], output_path)

# Move files for testing datasets
for dataset in test_datasets:
    move_files(dataset, dataset.split('_')[-1], output_path)
```
